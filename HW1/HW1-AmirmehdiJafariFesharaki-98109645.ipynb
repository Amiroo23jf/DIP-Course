## Question 1

The easiest acceptable way to select the value of the given pixel is bilinear interpolation using 4 nearest neighbors:

$$ f(x + a ,y + b) = \alpha a + \beta b + \gamma ab + \theta $$

$$ \theta = f(x + a, y + b)\Big{|}_{a,b=0} = f(x, y) $$

$$ f(x + 1, y) = \alpha + \theta \implies \alpha = f(x+1, y) - f(x, y)$$

$$ f(x, y + 1) = \beta + \theta \implies \beta = f(x, y+1) - f(x, y)$$

$$ f(x+1, y+1) = \alpha + \beta + \gamma + \theta \implies \gamma = f(x+1, y+1) - f(x+1,y) - f(x,y+1) + f(x,y) $$

combining the equations above, we find this final function for finding the value of the pixel:

$$ \boxed{f(x + a, y + b) = f(x,y) + \big[f(x+1, y) - f(x,y)\big] a + \big[f(x, y+1) - f(x,y)\big] b + \big[f(x+1, y+1) - f(x+1, y) - f(x, y+1) + f(x, y)\big] ab }$$

### Descriptions
The calculations above, show that the size of the uncompressed version of this image should be around <b>3566 KB</b>. However, the actual size of this image is only <b>143 KB</b> which is significantly lower than what we expected. The reason behind this is that PNG uses DEFLATE which is a type of <b>lossless</b> data compression algorithm using a combination of LZ77 and Huffman coding. That is why the actual size is unbelievably less than the size we calculated.

```python
im_actual_byte = 140341
compression_percentage = 100 - im_actual_byte / im_size_byte * 100
print("The Percentage of Compression is: "+"{:.2f}".format(compression_percentage)+"%")
```

This means that the compressed version's size is 3.84% of the size of the uncompressed version which is a huge save in storage.

## Question 3

### Part 1

We create the desired video in two phases. The first phase is to reach from RGB to Grayscale mode of the photo. For this, we first find the grayscale mode and then we reach it using a linearly. The algorithm is that imagine that for a selected pixel of the image, the intensity of that pixel in the grayscale equivalent is A, and the RGB intensities are R, G, and B respectively. Then in the i-th iteration from total n iterations, we set R = R + (A - R) * i / n and we do the same for G and B intensities. Using this algorithm we would finally reach the grayscale equivalent but in RGB mode in a smooth way.

```python
im = Image.open("bigMasoud.jpg")
im_gray = im.convert('L')
im_np = np.array(im)
im_np_gray = np.array(im_gray).astype("int16")
```

```python
im_list = []
n = 10
im_temp = im_np.copy()
im_red = im_np[:,:,0].astype("int16")
im_green = im_np[:,:,1].astype("int16")
im_blue = im_np[:,:,2].astype("int16")
for i in range(n+1):
    im_list.append(Image.fromarray(im_temp.astype("uint8")))
    im_temp[:,:,0] = np.round_(im_red + (im_np_gray - im_red) * i/n)
    im_temp[:,:,1] =  np.round_(im_green + (im_np_gray - im_green) * i/n)
    im_temp[:,:,2] = np.round_(im_blue + (im_np_gray - im_blue) * i/n)   
```

For now, we reached a grayscale image of "bigMasoud". Now we can continue with only the grayscale image and then we should enhance this image in order to reach the desired image. In order to do this, in the i-th iteration out of n iterations, we use the function below to map each internsity level to another intensity:

<center>
$s = \frac{e^{\beta (r-127.5)}}{1 + e^{\beta (r-127.5)}} (255 - (255-235)\frac{i}{n}) $ 
</center>

As it is shown, the function consists two parts. The first part is a simple shifted sigmoid. This sigmoid simply creates a contrast between the ligher and darker parts of the image in each iteration. In addition, the second part takes the responsibility of darkening the image over time. It simply maps the output of the sigmoid function ( which is a number between 0 to 1) to a number between 0 to (255 - (255 - 235) * i/n) linearly. This means that the maximum possible intensity is decreasing in each iteration which results in a darker image.

In this case, I tried different numbers for hyperparameters $\beta$ and n, and the values which result an acceptable quality were picked.

```python
im_arr = im_temp.copy()
beta = 1/55
n = 10
for i in range(n):
    im_list.append(Image.fromarray(im_arr.astype("uint8")))
    im_arr = np.round_(np.exp(beta * (im_arr.astype("int16") - 127.5))/(1 + np.exp(beta * (im_arr.astype("int16") - 127.5))) * (255 - (255 - 235) * i/n))
im_list[0].save("./results/Traumatized_bigMasoud.gif", save_all=True, append_images=im_list[1:], duration = 100)
```
